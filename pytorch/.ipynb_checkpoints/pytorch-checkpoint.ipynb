### What is tensor?
I mathematics, a tensor is an algebric object that describe a multilinear relationship between sets of algebraic objects related to a vector spasce. Objects that tensors may map between include vectos and scalars, and even other tensors.

In pytoch 

Tensors are a specialized data structure that are very similar to arrays and matrices. In PyTorch, we use tensors to encode the inputs and outputs of a model, as well as the model's parameters. Tensors are similar to NumPy's ndarrays, except that tensors can run on GPUs or other hardware accelerators.

#### The dot product is where we multiply matching members, then sum up:

(1, 2, 3) • (7, 9, 11)= 1×7 + 2×9 + 3×11
= 58
We match the 1st members (1 and 7), multiply them, likewise for the 2nd members (2 and 9) and the 3rd members (3 and 11), and finally sum them up.

```python
# Element wise multiplication 
tensor
```

```python
tensor*tensor
```

```python
# Matrix Multiplication 
torch.matmul(tensor,tensor) # 1*1 + 2*2 + 3*3 = 14
```

```python
%%time
value = 0 
for i in range(len(tensor)):
    value += tensor[i]*tensor[i]
print(value)
```

```python
%%time 
torch.matmul(tensor,tensor)
```

### There are two main rules that performing matrix multiplication needs to satisfy:

1. The **inner dimension** must match:
   * `(3,2) @ (3,2)` This won't work
   * `(2,3) @ (3,2)` This will work
   * `(3,2) @ (2,3)` This will work
2. The resulting matrix has the shape of the **outer dimensions**:
   * `(2,3) @ (3,2) -> (2,2)`
   * `(3,2) @ (2,3) -> (3,3)`

```python
torch.matmul(torch.rand(3,2),torch.rand(3,2)) # This will throws an error (Most Common Error)
```

```python
torch.matmul(torch.rand(2,3),torch.rand(3,2)) # This will  work
```

```python
torch.matmul(torch.rand(3,2),torch.rand(2,3)) # This will work
```

#### One of the most common errors in the deep learning: Shape Errors



```python
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_A.shape
```

```python
tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
tensor_B.shape
```

```python
torch.matmul(tensor_A,tensor_B)
```

#### Transpose --> we can manipulate the tensor shape issue 
A **transpose** switches the axes or dimensions of a given tensor.

```python
tensor_B.T
```

```python
tensor_B
```

```python
torch.matmul(tensor_A,tensor_B.T) # Transpose matches the inner dimension and this will not complain about the shape
```
